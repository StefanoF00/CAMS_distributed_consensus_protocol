# Eigenvalues of standard protocol

```python
import numpy as np
import matplotlib.pyplot as plt
import math
# Define eigenvalues of the Laplacian matrix for the example
lambdas = np.array([0.8, 0.8, 0.8])  # Use two eigenvalues for simplicity


# Smaller g
k_small = 0.1
mu_small = 1 - k_small * lambdas

# good g
k_good = 0.7
mu_good = 1 - k_good * lambdas

# Big g
k_big = 2.7
mu_big = 1 - k_big * lambdas
# Plotting
fig, ax = plt.subplots(figsize=(8, 8))
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.set_axisbelow(True)
# Plot unit circle
unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed')
ax.add_artist(unit_circle)

# Plot eigenvalues as points for small k
ax.scatter(mu_small[0], 0, color='red', label='$\\mu_1= 1 - k\\lambda_1$:  $k=0.1$;  $\\lambda_1=0.8+0i$')
ax.arrow(0, 0, mu_small[0]-0.05, 0, head_width=0.02, head_length=0.02, fc='red', ec='red',linewidth=3)

# Plot eigenvalues as points for good k
ax.scatter(-0.5*mu_good[1]-0.025, (math.sqrt(3)/2)*mu_good[1]+0.035, color='blue', label='$\\mu_2= 1 - k\\lambda_2$:  $k=0.7$;  $\\lambda_2=0.4+0.692i$')
ax.arrow(0, 0, -0.5*mu_good[1], (math.sqrt(3)/2)*mu_good[1], head_width=0.02, head_length=0.02, fc='blue', ec='blue',linewidth=3)

# Plot eigenvalues as points for small k
ax.scatter(0.5*mu_big[2]-0.025, (math.sqrt(3)/2)*mu_big[2]-0.035, color='green', label='$\\mu_3= 1 - k\\lambda_3$:  $k=2.7$;  $\\lambda_3=0.4-0.692i$')
ax.arrow(0, 0, 0.5*mu_big[2], (math.sqrt(3)/2)*mu_big[2], head_width=0.02, head_length=0.02, fc='green', ec='green',linewidth=3)

# Annotations


ax.annotate('$\\mu_1=1-0.1\\lambda_1$', (mu_small[0]-0.5, 0.15), color='red',fontsize=18)
ax.annotate('$\\mu_2=1-0.7\\lambda_2$', (-mu_good[1], mu_good[1]+0.1), color='blue',fontsize=18)
ax.annotate('$\\mu_3=1-2.7\\lambda_3$', (-mu_good[2]+0.1, -0.8), color='green',fontsize=18)

ax.set_xlabel('Real Part',fontsize=14)
ax.set_ylabel('Imaginary Part',fontsize=14)
ax.legend(fontsize=14, markerscale=1.5,loc='upper right')
ax.grid(True)
plt.xlim(-1.25, 1.25)
plt.ylim(-1.25, 1.25)
plt.gca().set_aspect('equal', adjustable='box')
plt.savefig('fig/Eigenvalues_std')
plt.show()
```

# Eigenvalues of New Centralized Protocol

```python
import numpy as np
import matplotlib.pyplot as plt
import math
# Define eigenvalues of the Laplacian matrix for the example
lambdas = np.array([0.8, 0.8, 0.8])  # Use two eigenvalues for simplicity

k = 70.0
# Smaller g
g_small = 0.45
mu_small = 1 - (k * lambdas)/(1+ k * g_small * lambdas)

# good g
g_good = 0.9
mu_good = 1 - (k * lambdas)/(1+ k * g_good * lambdas)

# Big g
g_big = 2.5
mu_big = 1 - (k * lambdas)/(1+ k * g_big * lambdas)
# Plotting
fig, ax = plt.subplots(figsize=(8, 8))
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.set_axisbelow(True)
# Plot unit circle
unit_circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed')
ax.add_artist(unit_circle)

# Plot eigenvalues as points for small g
ax.scatter(mu_small[0], 0, color='red', label='$\\mu_1$: $g_1=0.45$;  $\\lambda_1=0.8+0i$')
ax.arrow(0, 0, mu_small[0]+0.03, 0, head_width=0.02, head_length=0.02, fc='red', ec='red',linewidth=3)

# Plot eigenvalues as points for good g
ax.scatter(-0.5*mu_good[1]+0.03, (math.sqrt(3)/2)*mu_good[1]-0.03, color='blue', label='$\\mu_2$: $g_2=0.9$;  $\\lambda_2=0.4-0.692i$')
ax.arrow(0, 0, -0.5*mu_good[1], (math.sqrt(3)/2)*mu_good[1], head_width=0.02, head_length=0.02, fc='blue', ec='blue',linewidth=3)

# Plot eigenvalues as points for small g
ax.scatter(0.5*mu_big[2]+0.02, (math.sqrt(3)/2)*mu_big[2]+0.03, color='green', label='$\\mu_3$:  $g_3=2.5$;  $\\lambda_3=0.4+0.692i$')
ax.arrow(0, 0, 0.5*mu_big[2], (math.sqrt(3)/2)*mu_big[2], head_width=0.02, head_length=0.02, fc='green', ec='green',linewidth=3)

# Annotations


ax.annotate('$\\mu_1=1-\\frac{1+\\kappa g_1 \\lambda_1}{\\kappa \\lambda_1}$', (mu_small[0], 0.15), color='red',fontsize=18)
ax.annotate('$\\mu_2=1-\\frac{1+\\kappa g_2 \\lambda_2}{\\kappa \\lambda_2}$', (0.15, mu_good[1]), color='blue',fontsize=18)
ax.annotate('$\\mu_3=1-\\frac{1+\\kappa g_3 \\lambda_3}{\\kappa \\lambda_3}$', (0.15, mu_big[2]), color='green',fontsize=18)

ax.set_xlabel('Real Part',fontsize=14)
ax.set_ylabel('Imaginary Part',fontsize=14)
ax.legend(fontsize=14, markerscale=1.5,loc='lower right')
ax.grid(True)
plt.xlim(-1.25, 1.25)
plt.ylim(-1.25, 1.25)
plt.gca().set_aspect('equal', adjustable='box')
plt.savefig('fig/Eigenvalues_new')
plt.show()
```
